```
def impute_ethnicity(cols):
    height = cols[1]
    ethnicity = cols[0]
    if pd.isnull(ethnicity):
        if (height>170.361):
            return 'African American'
        elif(height<162.740):
            return 'Asian'
        elif (height>167 and height<170):
            return 'Caucasian'
        elif (height>162.740 and height<166):
            return 'Hispanic'
        elif (height>166. and height<168):
                return 'Other/Unknown'
        else:
            return 'Native American'
                
    else:
        return ethnicity
```
